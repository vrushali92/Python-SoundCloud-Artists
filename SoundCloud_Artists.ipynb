```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
artists_df = pd.read_csv("data/SoundCloud_Artists.csv")
print("Before removing empty cells size is {}".format(artists_df.shape))
```

# Cleaning Empty cells

```python
artists_df.dropna(inplace = True)
print("After removing empty cells size is {}".format(artists_df.shape))
```

```python
artists_df["Plays_Count"].fillna(0, inplace = True)
artists_df["Likes_Count"].fillna(0, inplace = True)
artists_df["Comments_Count"].fillna(0, inplace = True)
```

# Cleaning Date Format

```python
artists_df.dropna(subset = ["Dates_Created"], inplace = True)
artists_df.shape
```

# Removing duplicates

```python
artists_df.drop_duplicates(inplace = True)
artists_df.shape
```

```python
artists_df.info()
```

# Converting datatypes

```python
new_artists_df = artists_df.convert_dtypes()
new_artists_df.info()
```

```python
new_artists_df["Plays_Count"] = new_artists_df["Plays_Count"].str.replace(" ","")
new_artists_df["Likes_Count"] = new_artists_df["Likes_Count"].str.replace(" ","")
new_artists_df["Comments_Count"] = new_artists_df["Comments_Count"].str.replace(" ","")
print(new_artists_df.dtypes)
new_artists_df
```

```python
new_artists_df.Plays_Count = pd.to_numeric(new_artists_df.Plays_Count, errors='coerce')
new_artists_df.Likes_Count = pd.to_numeric(new_artists_df.Likes_Count, errors='coerce')
new_artists_df.Comments_Count = pd.to_numeric(new_artists_df.Comments_Count, errors='coerce')
new_artists_df.dtypes
```

```python
new_artists_df.corr()
```

```python
new_artists_df["Genres"].value_counts().head(10)
```

# Top 10 Genres

```python
new_artists_df["Genres"].value_counts().head(10).plot(kind = "bar", color = "green")
```

```python
order = new_artists_df["Genres"].value_counts().head(10).index
sns.countplot(data = new_artists_df, x = "Genres", order = order)
plt.xticks(rotation = 45, ha = "right")
```

# Top 10 Artists

```python
top_artists_df = new_artists_df[(new_artists_df["Followers"] > 30000) & (new_artists_df["Following"] > 1500)].head(10)
top_artists_df.shape
```

```python
sns.catplot(data = top_artists_df,
            x = "Followers",
            y = "Following",
            kind = "swarm",
            hue = "Genres")
plt.xticks(rotation = 45)
```

# Top 15 Followers Artist

```python
top15_followers_df = new_artists_df[["Username", "Genres", "Followers", "Following"]].nlargest(15, "Followers")
top15_followers_df
#sns.countplot(data = top10_followers_df, x = "Username")
sns.catplot(data = top15_followers_df,
           x = "Following",
           y = "Followers",
           kind = "swarm",
           hue = "Genres")
plt.xticks(rotation = 45)
```

# Top 15 Likes Artists

```python
top15_Likes_df = new_artists_df[["Username", "Genres", "Followers", "Following", "Likes_Count"]].nlargest(15, "Likes_Count")
top15_Likes_df

sns.catplot(data = top15_Likes_df,
            y = "Likes_Count",
            x = "Followers",
            kind = "strip",
            hue = "Genres")
plt.xticks(rotation = 45)
```

# Top 15 Plays Count

```python
top15_Plays_df = new_artists_df[["Username", "Genres", "Followers", "Likes_Count", "Plays_Count"]].nlargest(15, "Plays_Count")
top15_Plays_df
sns.catplot(data = top15_Plays_df,
           y = "Plays_Count",
           x = "Followers",
           kind = "strip",
           hue = "Genres")
plt.xticks(rotation = 45)
```

# Followers vs Following vs Plays_Count

```python
top15_Plays_df = new_artists_df[["Username", "Genres", "Followers", "Following", "Plays_Count"]].nlargest(15, "Plays_Count")
sns.pairplot(top15_Plays_df, hue = "Genres", height = 2, kind = "scatter")
```
